# Factorial
# Fibonacci


```python
def fact(n):
    if n==0 or n==1:
        return 1
    else:
        return n*fact(n-1)

print(fact(4))
```

```python
def fibo(n):
    if n==0:
        return 0
    elif n==1:
        return 1
    else :
        return fibo(n-1)+fibo(n-2)
    
print(fibo(40))
```

```python
def fibo(n):
    F=[]
    F.append(0)
    F.append(1)
    for i in range (2,n+1):
        F.append(F[i-1]+F[i-2])
    return F[-1]
    
print(fibo(100))
```

```python
def linespace(l,u,n):
    a=[]
    a.append(l)
    h=(u-l)/n
    for i in range(1,n):
        
        a.append(a[i-1]+h)
    a.append(u)
    return a

print(linespace(-5,5,100))
        
```
